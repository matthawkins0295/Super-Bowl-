```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
url = 'https://en.wikipedia.org/wiki/List_of_Super_Bowl_champions'
```

```python
page = requests.get(url)
page
```

```python
soup1 = BeautifulSoup(page.text, 'html')
```

```python
print(soup1.prettify())
```

```python
soup1.find('table')
```

```python
soup1.find_all('table', class_ = "wikitable sortable")
```

```python
soup1.find('table', class_ = "wikitable sortable").text
```

```python
soup1.find('table', class_ = "wikitable sortable").text.strip()
```

```python
soup1.find_all('table')[1]
```

```python
table = soup1.find_all('table')[1]
```

```python
print(table)
```

```python
super_bowl = table.find_all('th')
```

```python
super_bowl
```

```python
super_bowl_champs = [title.text.strip() for title in super_bowl]
```

```python
print(super_bowl_champs)
```

```python
df = pd.DataFrame(columns = super_bowl_champs)
```

```python
column_data = table.find_all('tr')
```

```python
for row in column_data[1:]:
    row_data = row.find_all('td')
    individual_row_data = [data.text.strip() for data in row_data]

    length = len(df)
    df.loc[length] = individual_row_data 
```

```python
df = df.drop('Ref', axis=1)
```
